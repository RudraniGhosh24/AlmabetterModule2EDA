## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

try:

    # Create copies so raw data remains unchanged
    titles_clean = titles_df.copy()
    credits_clean = credits_df.copy()

    # Remove duplicate records
    titles_clean.drop_duplicates(inplace=True)
    credits_clean.drop_duplicates(inplace=True)

    # Handle missing values
    titles_clean["age_certification"] = (titles_clean["age_certification"].fillna("Unknown"))

    # Median imputation for ratings
    for col in ["imdb_score","tmdb_score","runtime"]:titles_clean[col].fillna(titles_clean[col].median(),inplace=True)

    # Seasons missing for movies is logical
    titles_clean["seasons"] = (titles_clean["seasons"].fillna(0))


    # Feature Engineering

    # Primary genre extraction
    titles_clean["main_genre"] = (titles_clean["genres"].astype(str).str.strip("[]").str.split(",").str[0])


    # Primary production country
    titles_clean["main_country"] = (titles_clean["production_countries"].astype(str).str.strip("[]").str.split(",").str[0])


    # Content age
    current_year=2024

    titles_clean["content_age"] = (current_year- titles_clean["release_year"])


    # Merge title and credits for contributor analysis
    merged_df=pd.merge(titles_clean, credits_clean, on="id", how="left")

    print("Data wrangling completed successfully.")

except Exception as e:
    print(e)
````

### What all manipulations have you done and insights you found?



```
The following manipulations were performed:

- Removed duplicate records from both datasets
- Handled missing values using:
  median imputation for numerical variables
  logical imputation (0) for seasons
  “Unknown” for missing certifications
- Extracted primary genre and production country from multi-valued fields.
- Created derived variable content_age for temporal analysis.
- Merged titles and credits datasets for actor/director-level insights.

Insights Found
- Missing values in seasons were structural rather than data errors.
- Genres and countries needed transformation before meaningful analysis.
- Feature engineering made the data suitable for UBM-based exploration.```

Explain Briefly.

Based on the comprehensive EDA of Amazon Prime Video's catalog of 9,000+ titles and 124,000+ credits, the following data-driven strategies are recommended to achieve the platform's core business objectives of content acquisition excellence, subscriber growth, and long-term retention.

1. Rebalance the Content Mix Toward TV Shows
TV Shows demonstrated consistently higher median IMDb scores and stronger episodic retention mechanics than movies. Amazon Prime should increase the share of multi-season series — particularly in high-performing genres like Drama and Thriller — to reduce monthly churn. The data shows that 1-season dominance limits the 'binge-anchor' effect that retains subscribers month-over-month.

2. Invest in Genre Diversification — Fill the White Space
Drama, Comedy, and Action dominate the catalog while Horror, Sci-Fi, Documentary, and International genres are underrepresented. These niche genres have highly loyal, underserved audiences. Targeted acquisition in these categories can grow subscriber segments without cannibalizing existing Drama/Comedy viewership.

3. Develop Family-Friendly Content as a Strategic Priority
The Age Certification analysis revealed that adult-oriented content dominates while family/children content is severely underrepresented. Families are a high-value, low-churn subscriber segment. Increasing G/PG-rated original content would directly compete with Disney+ in the family segment — currently ceded almost entirely to competitors.

4. Build a Quality-Weighted Recommendation Engine
The correlation analysis showed that IMDb score, TMDb score, and IMDb vote count are meaningfully related but not redundant. A composite content quality score combining all three metrics (with vote count as a confidence weight) would produce more reliable content recommendations than any single metric alone — directly improving session satisfaction scores.

5. Internationalize the Content Library Aggressively
The production country analysis showed heavy US and India dominance. Latin America, Southeast Asia, and Africa are fast-growing streaming markets with extremely limited local-language representation in Prime's catalog. Investing in regional originals and dubbing/subtitling for these markets would accelerate international subscriber acquisition at a lower cost than US-market growth.

6. Prune Low-Engagement Legacy Content & Improve Discoverability
A significant portion of the catalog consists of low-vote, legacy-era titles (1940s–1970s) with below-average IMDb scores. These bulk up title count without contributing to genuine viewer engagement. Periodic catalog audits to deprioritize or remove very low-rated, low-engagement titles would improve the average quality perception and make high-quality content more discoverable.
